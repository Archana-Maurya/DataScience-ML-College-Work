```python
#Capstone Project: Student Success & Career Path Prediction Scenario

#The university wants to analyze student performance data to:

# 1.Predict exam scores (Regression).
# 2.Classify students into “At Risk” vs. “On Track” categories (Classification).
# 3.Cluster students into groups with similar study habits (Clustering).
# 4.Recommend interventions (extra tutoring, workshops, counseling).

# Student Success Capstone Project


#step 1: import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import r2_score, mean_squared_error, classification_report
from sklearn.cluster import KMeans

# Step2: Load dataset
df = pd.read_csv("Student Success & Career Path  - Sheet1.csv")

# Data Preprocessing
# Remove Student_ID (not useful for ML)
df = df.drop("Student_ID", axis=1)

# Encode Gender
df["Gender"] = LabelEncoder().fit_transform(df["Gender"])

# Encode Pass_Fail
df["Pass_Fail"] = LabelEncoder().fit_transform(df["Pass_Fail"])


# Step3: REGRESSION
# Predict Final_Exam_Score
X_reg = df.drop("Final_Exam_Score", axis=1)
y_reg = df["Final_Exam_Score"]

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg, y_reg, test_size=0.2, random_state=42
)

reg_model = LinearRegression()
reg_model.fit(X_train_reg, y_train_reg)

y_pred_reg = reg_model.predict(X_test_reg)

print("Regression R2 Score:", r2_score(y_test_reg, y_pred_reg))
print("Regression RMSE:", np.sqrt(mean_squared_error(y_test_reg, y_pred_reg)))


# step4: CLASSIFICATION
# Predict Pass_Fail
X_clf = df.drop("Pass_Fail", axis=1)
y_clf = df["Pass_Fail"]

X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
    X_clf, y_clf, test_size=0.2, random_state=42
)

scaler = StandardScaler()
X_train_clf_scaled = scaler.fit_transform(X_train_clf)
X_test_clf_scaled = scaler.transform(X_test_clf)

clf_model = LogisticRegression()
clf_model.fit(X_train_clf_scaled, y_train_clf)

y_pred_clf = clf_model.predict(X_test_clf_scaled)

print("\nClassification Report:")
print(classification_report(y_test_clf, y_pred_clf))


# Step5: CLUSTERING
# Group Students
cluster_features = df[[
    "Hours_Studied",
    "Attendance (%)",
    "Final_Exam_Score"
]]

kmeans = KMeans(n_clusters=3, random_state=42)
df["Cluster"] = kmeans.fit_predict(cluster_features)

print("\nCluster Means:")
print(df.groupby("Cluster").mean())


# Step6: RECOMMENDATION SYSTEM
def recommend(row):
    if row["Cluster"] == 0:
        return "Extra Tutoring"
    elif row["Pass_Fail"] == 0:
        return "Counseling"
    else:
        return "Career Workshop"

df["Recommendation"] = df.apply(recommend, axis=1)

print("\nFinal Output Sample:")
print(df[["Final_Exam_Score", "Pass_Fail", "Cluster", "Recommendation"]].head())
```
